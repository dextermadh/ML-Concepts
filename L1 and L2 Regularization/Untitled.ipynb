```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns 
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.shape 
```

```python
df.isnull().sum() 
```

```python
df.nunique()
```

```python
df.drop(['Date', 'Address', 'Postcode', 'YearBuilt', 'Lattitude', 'Longtitude'], axis='columns', inplace=True) 
df.head() 
```

```python
df.shape 
```

```python
df.isnull().sum()
```

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] =  df[cols_to_fill_zero].fillna(0)
df 
```

```python
df.isnull().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean()) 
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean()) 
df.isna().sum()
```

```python
df.dropna(inplace=True) 
df.isna().sum()
```

```python
df.head() 
```

```python
dummies = pd.get_dummies(df[['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname']], dtype="int64", drop_first=True)     
dummies.head()  
```

```python
df.drop(['Suburb', 'Type', 'Method', 'SellerG', 'CouncilArea', 'Regionname'], axis='columns', inplace=True)
df = pd.concat([df, dummies], axis='columns') 
df.head() 
```

```python
X = df.drop('Price', axis='columns') 
y = df.Price 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression() 
```

```python
model.fit(X_train, y_train) 
```

```python
model.score(X_test, y_test) 
```

```python
model.score(X_train, y_train) 
```

```python
# model is overfitting because it is working well with training dataset and not with test dataset
```

### Lasso Regularization -L1 

```python
from sklearn import linear_model 

lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1) 

lasso_reg.fit(X_train, y_train) 
```

```python
lasso_reg.score(X_test, y_test) 
```

```python
lasso_reg.score(X_train, y_train) 
```

### Ridge Regression - L2 

```python
from sklearn.linear_model import Ridge 

ridge_reg = Ridge(alpha=50, max_iter=100, tol=0.1) 
ridge_reg.fit(X_train, y_train) 
```

```python
ridge_reg.score(X_test, y_test) 
```
