```python
def z_score_outlier_remover(x, df): 
    df[f'zscore_{x}'] = (df[x] - df[x].mean()) / df[x].std()
    df2 = df[(df[f'zscore_{x}'] > -3) & (df[f'zscore_{x}'] < 3)]
    return df2
```

```python
columns = ['Age', 'RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak'] 
for i in columns: 
    df2 = z_score_outlier_remover(i, df)    

drop = ['zscore_Age', 'zscore_RestingBP', 'zscore_Cholesterol', 'zscore_FastingBS', 'zscore_MaxHR', 'zscore_Oldpeak']
df2.drop(drop, axis='columns', inplace=True)     
df2.head() 
```

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler(feature_range=(0,1))
X_scaled = scaler.fit_transform(X)
X_scaled 
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30) 
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier 
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto'),   
        'params': {
            'C': [1,10,20], 
            'kernel': ['rbf', 'linear']
        }
    }, 
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'), 
        'params': {
            'C': [1,5,10]
        }
    },   
    'random_forest': {
        'model': RandomForestClassifier(), 
        'params': {
            'n_estimators': [1,5,10]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores = [] 

for model_name, mp in model_params.items(): 
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False) 
    clf.fit(X_scaled, y) 
    scores.append({
        'model': model_name, 
        'best_score': clf.best_score_, 
        'best_params': clf.best_params_
    })
```

```python
score_df = pd.DataFrame(scores) 
score_df 
```

```python
from sklearn.decomposition import PCA 

pca = PCA(0.95) 

X_pca = pca.fit_transform(X_scaled) 
X_pca.shape 
```

```python
scores = [] 

for model_name, mp in model_params.items(): 
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False) 
    clf.fit(X_pca, y) 
    scores.append({
        'model': model_name, 
        'best_score': clf.best_score_, 
        'best_params': clf.best_params_
    })

score_df = pd.DataFrame(scores) 
score_df 
```
