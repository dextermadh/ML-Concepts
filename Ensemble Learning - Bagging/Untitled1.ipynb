```python
def z_score_outlier_removal(x, df):  
    df[f'zscore_{x}'] = (df[x] - df[x].mean()) / df[x].std()
    df2 = df[(df[f'zscore_{x}'] > -3) & (df[f'zscore_{x}'] < 3)]
    return df2 
```

```python
columns = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

for column in columns: 
    df2 = z_score_outlier_removal(column, df)

df2.head() 
```

```python
df2.shape 
```

```python
drop_columns = ['zscore_Age', 'zscore_RestingBP', 'zscore_Cholesterol', 'zscore_MaxHR', 'zscore_Oldpeak']
df2.drop(drop_columns, axis='columns', inplace=True)  
df2.head() 
```

```python
from sklearn.preprocessing import LabelEncoder 

le = LabelEncoder() 
dfle = df2 
```

```python
def lable_encode(x, df): 
    df[x] = le.fit_transform(df[x]) 
```

```python
columns = ['Sex', 'ChestPainType', 'ExerciseAngina', 'ST_Slope'] 

for column in columns: 
    lable_encode(column, dfle) 

dfle.head() 
```

```python
dummies = pd.get_dummies(dfle[['RestingECG']], dtype='int64')
dummies.head() 
```

```python
df3 = pd.concat([dfle, dummies], axis='columns')
df3.shape 
```

```python
df3.drop('RestingECG', axis='columns', inplace=True) 
```

```python
df3.isnull().sum()
```

```python
X = df3.drop('HeartDisease', axis='columns')
y = df3.HeartDisease 
```

```python
df3.head() 
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
X_scaled = scaler.fit_transform(X)
X_scaled 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC 
model = SVC(C=10, kernel='rbf', gamma='auto')
model.fit(X_train, y_train) 
```

```python
model.score(X_test, y_test) 
```

```python
from sklearn.ensemble import BaggingClassifier 
```

```python
bag_model = BaggingClassifier(
    estimator=SVC(), 
    n_estimators=50, 
    max_samples=0.8,  
    oob_score=True,
    random_state=0 
)

bag_model.fit(X_train, y_train) 
```

```python
bag_model.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier 
```

```python
model = DecisionTreeClassifier() 
model.fit(X_train, y_train) 
```

```python
model.score(X_test, y_test) 
```

```python
bag_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(), 
    n_estimators=100, 
    max_samples=0.8, 
    oob_score=True, 
    random_state=0
)
bag_model.fit(X_train, y_train) 
```

```python
bag_model.score(X_test, y_test) 
```
